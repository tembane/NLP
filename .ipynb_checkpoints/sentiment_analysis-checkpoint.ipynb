```python
import numpy as np
import pandas as pd
import nltk
import matplotlib.pyplot as plt
from scipy.special import softmax
from nltk import sent_tokenize, word_tokenize

file_name = 'data.csv'
MODEL = f"cardiffnlp/twitter-roberta-base-sentiment-latest"

```

```python
df = pd.read_csv(file_name)
eng_comments = df.loc[df.Key == 'en', 'Value'].tolist()
```

```python
sample = []
for comment in eng_comments:
    if len(sent_tokenize(comment)) == 1:
        sample.append(comment)
```

```python
#график кол-ва слов в комментарии
nltk.download('punkt')
word_counts = [len(word_tokenize(comment)) for comment in sample]
comm_len_df = pd.DataFrame({'Word count': word_counts})
plt.hist(comm_len_df['Word count'], bins=max(comm_len_df['Word count']))
plt.xticks()
plt.yticks()
plt.xlabel('Количество слов в комментарии')
plt.ylabel('Количество комментариев')
plt.show()
```

```python
from transformers import AutoModelForSequenceClassification
from transformers import AutoTokenizer, AutoConfig

tokenizer = AutoTokenizer.from_pretrained(MODEL)
config = AutoConfig.from_pretrained(MODEL)
model = AutoModelForSequenceClassification.from_pretrained(MODEL)

encoded_input = [tokenizer.encode_plus(comment, return_tensors='pt', padding='max_length', truncation=True, max_length=40) for comment in sample]
print(len(encoded_input))
output = [model(**encoded_input[i]) for i in range(len(encoded_input))]
print(len(output))
scores = [output[i][0][0].detach().numpy() for i in range(len(output))]
scores = softmax(scores)
print(len(scores))
```
